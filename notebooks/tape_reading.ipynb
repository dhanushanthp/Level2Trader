```python
data = pd.read_csv('test_data/2021040123_ACY.csv', names=['time', 'bid', 'bid_size', 'ask', 'ask_size', 'last', 'size', 'type'])
```

```python
from tape_reading.tape_reader import TapeReader
```

```python
tr = TapeReader('tick', False)
```

```python
from util.price_util import PriceUtil
```

```python
data['closest'] = data.apply(lambda x: tr.find_closest(x['bid'], x['ask'], x['last']), axis=1)
```

```python
data['price_on'] = data.apply(lambda x: 'bid' if x['closest'] == x['bid'] else 'ask', axis=1)
```

```python
data['size_'] = data['size'].apply(lambda x: round(x/100))
```

```python
data['ask_size'] = data['ask_size']/100
data['bid_size'] = data['bid_size']/100
```

```python
time_range = [f'23:33:{str(i).zfill(2)}' for i in range(17, 23)]
filtered_data = data[data['time'].isin(time_range)]
```

```python
# filtered_data = data
```

```python
agg = filtered_data.groupby(['time','price_on','closest']).agg({'size_':['sum'], 
                                                          'bid_size':['last'], 'ask_size':['last']}).reset_index()
```

```python
agg.columns = ['time', 'type', 'close_price', 'size', 'bid_size', 'ask_size']
```

```python
agg['output_str'] = agg.apply(lambda x: f'{x["size"]}/{x["ask_size"]}' if x['type'] is 'ask' else f'{x["size"]}/{x["bid_size"]}', axis=1)
```
